#  1.)  For the suburb Altona, it is postulated that a typical property sells for 800,000. Use the data at hand to test this assumption. Is the typical property price really 800,000 or has it increased? Use a significance level of 5%. 

```python
import pandas as pd
import numpy as np

# Filter Altona properties
altona_prices = df[df['Suburb'] == 'Altona']['Price']

# Hypothesized mean
mu = 800000

# Calculate sample statistics using pandas
n = altona_prices.count()
sample_mean = altona_prices.mean()
sample_std = altona_prices.std(ddof=1)  # Sample standard deviation

# Calculate the t-statistic using pandas calculations
t_stat = (sample_mean - mu) / (sample_std / np.sqrt(n))

# Calculate the two-tailed p-value using the t-distribution
from scipy.stats import t
p_value = 2 * (1 - t.cdf(abs(t_stat), df=n-1))

# Display the results
print("Sample Size:", n)
print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std)
print("T-Statistic:", t_stat)
print("P-Value:", p_value)
```

The p-value (0.307) confirms that we fail to reject the null hypothesis. The typical property price in Altona is not significantly different from $800,000 at the 5% significance level. 

# 2 . For the year 2016, is there any difference in prices of properties sold in the summer months vs winter months? Consider months from October till March as winter months and the rest as summer months. Use a significance level of 5%. 

```python
# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')

# Filter properties sold in 2016
data_2016 = df[df['Date'].dt.year == 2016].copy()  # Use .copy() to avoid the warning

# Extract month from Date
data_2016['Month'] = data_2016['Date'].dt.month

# Define winter (Oct-Mar) and summer (Apr-Sep) months
winter_months = [10, 11, 12, 1, 2, 3]
summer_months = [4, 5, 6, 7, 8, 9]

# Split prices into winter and summer groups
winter_prices = data_2016[data_2016['Month'].isin(winter_months)]['Price']
summer_prices = data_2016[data_2016['Month'].isin(summer_months)]['Price']

# Perform two-sample t-test
from scipy.stats import ttest_ind
t_stat, p_value = ttest_ind(winter_prices, summer_prices, equal_var=False)  # Welch's t-test

# Display results
print("T-Statistic:", t_stat)
print("P-Value:", p_value)

# Check significance
if p_value < 0.05:
    print("Conclusion: Significant difference in prices (reject H0).")
else:
    print("Conclusion: No significant difference in prices (fail to reject H0).")
```

# 3 .For the suburb of Abbotsford, what is the probability that out of 10 properties sold, 3 will not have car parking? Use the column car in the dataset. Round off your answer to 3 decimal places. 

```python
from math import comb

# Filter properties in Abbotsford
abbotsford_data = df[df['Suburb'] == 'Abbotsford']

# Total properties in Abbotsford
total_properties_abbotsford = abbotsford_data.shape[0]

# Properties without car parking
properties_without_car = abbotsford_data[abbotsford_data['Car'] == 0].shape[0]

# Calculate p (probability of no car parking)
p_no_car = properties_without_car / total_properties_abbotsford

# Binomial calculation for P(X = 3) when n = 10 and k = 3
n = 10
k = 3
binomial_probability = comb(n, k) * (p_no_car**k) * ((1 - p_no_car)**(n - k))

# Round off to 3 decimal places
binomial_probability_rounded = round(binomial_probability, 3)

print("Probability of no car parking (p):", round(p_no_car, 3))
print("Probability of exactly 3 out of 10 properties without car parking:", binomial_probability_rounded)
```

# In the suburb of Abbotsford, what are the chances of finding a property with 2 bathrooms? Round your answer to 3 decimal places. 

```python
# Filter properties in Abbotsford
abbotsford_data = df[df['Suburb'] == 'Abbotsford']

# Total properties in Abbotsford
total_properties_abbotsford = abbotsford_data.shape[0]

# Properties with 3 rooms in Abbotsford
properties_with_3_rooms = abbotsford_data[abbotsford_data['Rooms'] == 3].shape[0]

# Calculate probability
probability = properties_with_3_rooms / total_properties_abbotsford

# Round to 3 decimal places
probability_rounded = round(probability, 3)

print("Total Properties in Abbotsford:", total_properties_abbotsford)
print("Properties with 3 Rooms:", properties_with_3_rooms)
print("Probability of 3 Rooms:", probability_rounded)
```
